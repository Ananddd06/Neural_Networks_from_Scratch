```python
import numpy as np

# -----------------------------
# Hyperparameters
# -----------------------------
np.random.seed(42)
N = 32          # batch size
D = 100         # input features
H1, H2, H3 = 64, 32, 16
O = 2
lr = 0.01
epochs = 100

# -----------------------------
# Data
# -----------------------------
X = np.random.randn(N, D)
Y = np.random.randn(N, O)

# -----------------------------
# Initialize weights & biases
# -----------------------------
W1 = 0.01 * np.random.randn(D, H1)
b1 = np.zeros((1, H1))

W2 = 0.01 * np.random.randn(H1, H2)
b2 = np.zeros((1, H2))

W3 = 0.01 * np.random.randn(H2, H3)
b3 = np.zeros((1, H3))

W4 = 0.01 * np.random.randn(H3, O)
b4 = np.zeros((1, O))

# -----------------------------
# Training loop
# -----------------------------
for epoch in range(1, epochs + 1):

    # -------- Forward pass --------
    Z1 = X @ W1 + b1
    A1 = np.maximum(0, Z1)

    Z2 = A1 @ W2 + b2
    A2 = np.maximum(0, Z2)

    Z3 = A2 @ W3 + b3
    A3 = np.maximum(0, Z3)

    Z4 = A3 @ W4 + b4
    Y_hat = Z4

    # -------- Loss (MSE) --------
    loss = np.mean((Y_hat - Y) ** 2)

    # -------- Backprop --------
    dY_hat = 2 * (Y_hat - Y) / N

    dW4 = A3.T @ dY_hat
    db4 = np.sum(dY_hat, axis=0, keepdims=True)
    dA3 = dY_hat @ W4.T

    dZ3 = dA3 * (Z3 > 0)
    dW3 = A2.T @ dZ3
    db3 = np.sum(dZ3, axis=0, keepdims=True)
    dA2 = dZ3 @ W3.T

    dZ2 = dA2 * (Z2 > 0)
    dW2 = A1.T @ dZ2
    db2 = np.sum(dZ2, axis=0, keepdims=True)
    dA1 = dZ2 @ W2.T

    dZ1 = dA1 * (Z1 > 0)
    dW1 = X.T @ dZ1
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # -------- Update --------
    W4 -= lr * dW4
    b4 -= lr * db4

    W3 -= lr * dW3
    b3 -= lr * db3

    W2 -= lr * dW2
    b2 -= lr * db2

    W1 -= lr * dW1
    b1 -= lr * db1

    # -------- Print loss --------
    print(f"Epoch {epoch:3d} | Loss: {loss:.6f}")

# -----------------------------
# Final weights & biases
# -----------------------------
print("\n===== UPDATED PARAMETERS AFTER 100 ITERATIONS =====")

print("\nW1:\n", W1)
print("\nb1:\n", b1)

print("\nW2:\n", W2)
print("\nb2:\n", b2)

print("\nW3:\n", W3)
print("\nb3:\n", b3)

print("\nW4:\n", W4)
print("\nb4:\n", b4)
```

$$
